```python
#Loading python dependents for the project
import numpy as np # linear algebra
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   
import os
```

```python
#defining file name
filename='SampleSuperstore.csv'
```

```python
# reading the file
df = pd.read_csv(os.path.join(os.getcwd(),filename))
```

```python
#display file contents for our refernce that correct file is loaded
# head gives the top few contents with headers 
print('Displaying file headers and contents\n')
df.head()
```

```python
# Print the column names  and also check if any column has null values  and also see how many rows or data entry we have in the data 
# if there are null values we may have to see how to fill the values or we need to drop that row . If no null values then 
# we are good to continue 
print('Printing Column headers name and checking for null value \n {} '.format(df.isnull().sum()))
print('\nPrinting data type \n {} '.format(df.dtypes))
num=len(df)
print('\n Total Entries in the data is {} '.format(num))

```

```python
#  Describe the data to get an overview
df.describe()
 
```

```python
# We need to drop any duplicate items if exist in the dataset 
print("Total number of duplicate items in the data set {}".format(df.duplicated().sum()))
#drop duplicate data if there is any 
if df.duplicated().sum() > 0:
    new_data=df.drop_duplicates()
    print("Total number of duplicate items in the data set after dropping {}".format(new_data.duplicated().sum()))
# find unique contents in dataset
new_data.nunique()
```

```python
# We need to check the correlation in data set 
new_data.corr()
# Row and column  should be cross checked , example postal code vs postal code is 1 means there is higher correlation . But we should not look at diagonal
# item as its corelation among same data .In the matrix if u see  profit has a highest corelation with Quantity and Quantity has highest corelation with sales 
# while sales with Profit . Discount and sales have a least corelation 
```

```python
# checking covariance of data 
new_data.cov()

# The coefficient ranges from negetive  to positive ollowing:
#Positive correlation — tells us that both variables move in the same direction e.g. both values are increasing simultaneously.
#Negative correlation — describes inversely correlated variables. Meaning, if one variable is increasing the other is decreasing or vice-versa.
#A correlation coefficient of zero shows that there is no relationship at all.
```

```python
# delete columns which do not imply anything for analysis
col=['Postal Code']

data=new_data.drop(columns=col,axis=1)
```

```python
# state wise distribution of information 
plt.figure(figsize=(15,8))
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(15,8))
sns.countplot(x=data['Sub-Category'])
plt.xticks(rotation=90)
plt.show()

sns.countplot(x=data['Segment'])
sns.countplot(x=data['Region'])
sns.countplot(x=data['Country'])
```

```python
# from above result we can infer one thing is  country is common so we can drop the colum 
col=['Country']

data1=data.drop(columns=col,axis=1)
```

```python
data_grouped=pd.DataFrame(data1.groupby(['Ship Mode','Segment','Category','Sub-Category','State','Region'])['Quantity','Discount','Sales','Profit'].sum().reset_index())
data_grouped
```

```python
fig, ax = plt.subplots(figsize = (10 , 6))
ax.scatter(data_grouped["Sales"] , data_grouped["Profit"])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
print(data_grouped['Sales'].describe())
plt.figure(figsize = (9 , 8))
sns.distplot(data_grouped['Sales'], color = 'b', bins = 100, hist_kws = {'alpha': 0.4});
```

```python
print('\nPrinting data type \n {} '.format(data_grouped.dtypes))
```

```python
#01-----------   PRODUCT LEVEL ANALYSIS
# listing down product categories and subcategories 
print('--- CATEGORIES DETAILS ---  \n{} \n'.format(data_grouped['Category'].value_counts()))
print('--- SUB CATEGORIES DETAILS --- \n {}\n'.format(data_grouped['Sub-Category'].value_counts()))    

print('WE KNOW THAT THERE ARE ONLY 2 CATEGORIES , NOW LETS SEE WHICH SUB CATEGORY IS DISTRIBUTE FOR CATEGORY ')
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data_grouped,color='y')
plt.show()
print('Phones , Accesories ,Machiners and Copiers are falling under technology rest everything under Furniture ')

plt.figure(figsize=(12,10))
data_grouped['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
print('The Data shows  Office Supplies especially in Binders and Paper Sub category have high percentage sales \n, hence lets look at the profit \n ')
data_grouped.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()
print("Highest profit is earned in Copiers while Selling price for Chairs and Phones is extremely high \n compared to other products. Another interesting fact- people dont prefer to buy Tables and \n Bookcases from Superstore. Hence these departments are in loss")

print(' Presenting Region based data ')
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=data_grouped)
plt.show()

data_grouped['Cost']=data_grouped['Sales']-data_grouped['Profit']
data_grouped['Cost'].head()
data_grouped['Profit %']=(data_grouped['Profit']/data_grouped['Cost'])*100
```

```python
data_grouped
```

```python
# Segregate data when profit is less than zero 
data_grouped_less_than_zero=data_grouped[data_grouped['Profit %'] < 0]
data_grouped_less_than_zero
```

```python
# Segregate data when profit is greater than zero but less than 50%
data_grouped_greater_than_zero=data_grouped[data_grouped['Profit %'] > 0]
data_grouped_zero_50percent=data_grouped_greater_than_zero[data_grouped_greater_than_zero['Profit %'] < 50]
data_grouped_zero_50percent
```

```python
data_grouped_zero_50percent.sort_values(['Profit %','Sub-Category'],ascending=False).groupby('Profit %').head(5)
```

```python
# Segregate data when profit is greater  50%
data_grouped_zero_50percent_greater=data_grouped[data_grouped['Profit %'] > 50]
data_grouped_zero_50percent_greater
data_grouped_zero_50percent_greater.sort_values(['Profit %','Sub-Category'],ascending=False).groupby('Profit %').head(5)
```

```python
 
 
print('--- CATEGORIES DETAILS ---  \n{} \n'.format(data_grouped_zero_50percent['Category'].value_counts()))
print('--- SUB CATEGORIES DETAILS --- \n {}\n'.format(data_grouped_zero_50percent['Sub-Category'].value_counts()))    


plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data_grouped_zero_50percent,color='r')
plt.show()


plt.figure(figsize=(12,10))
data_grouped_zero_50percent['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()

data_grouped_zero_50percent.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()


print(' Presenting Region based data ')
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=data_grouped_zero_50percent)
plt.show()

```

```python
print('--- CATEGORIES DETAILS ---  \n{} \n'.format(data_grouped_zero_50percent_greater['Category'].value_counts()))
print('--- SUB CATEGORIES DETAILS --- \n {}\n'.format(data_grouped_zero_50percent_greater['Sub-Category'].value_counts()))    


plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data_grouped_zero_50percent_greater,color='r')
plt.show()


plt.figure(figsize=(12,10))
data_grouped_zero_50percent_greater['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()

data_grouped_zero_50percent_greater.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()


print(' Presenting Region based data ')
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=data_grouped_zero_50percent_greater)
plt.show()

```
